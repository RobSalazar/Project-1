```python
%matplotlib notebook
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
File=pd.read_csv("../clean_farm.csv")
File
File.fillna("N/A", axis=0, inplace=True)

NewFile=File[File['city'].isin([ "New York","San Francisco", "Los Angeles", "Seattle", "Boston"])]
```

```python
Products=NewFile.set_index('city')
```

```python
Products.columns
```

```python
Products
```

```python
df=Products[['Organic', 'Bakedgoods',
       'Cheese', 'Crafts', 'Flowers', 'Eggs', 'Seafood', 'Herbs', 'Vegetables',
       'Honey', 'Jams', 'Maple', 'Meat', 'Nursery', 'Nuts', 'Plants',
       'Poultry', 'Prepared', 'Soap', 'Trees', 'Wine', 'Coffee', 'Beans',
       'Fruits', 'Grains', 'Juices', 'Mushrooms', 'PetFood', 'Tofu',
       'WildHarvested']]
df
```

```python
ProductFile=df.reset_index()
ProductFile
```

```python
Prodc_df=ProductFile.replace(["-", "Y", "N", "N/A"], [0, 1, 0, 0])
```

```python
Prodc_df
```

```python
Sum_df=Prodc_df.groupby('city').sum()
Sum_df
```

```python
Total_Products=Sum_df.loc[:,'Total Product'] = Sum_df.sum(axis=1)
Total_Products
```

```python
Sum_df
```

```python
NewTotal=Sum_df.reset_index()
NewTotal
```

```python
NewTotal.plot(kind="bar", x="city", y="Total Product", grid=True, figsize=(8,8),
              title="Total Veriety Products")
```

```python
Chart_df=NewTotal[['city','Organic', 'Bakedgoods',
       'Cheese', 'Crafts', 'Flowers', 'Eggs', 'Seafood', 'Herbs', 'Vegetables',
       'Honey', 'Jams', 'Maple', 'Meat', 'Nursery', 'Nuts', 'Plants',
       'Poultry', 'Prepared', 'Soap', 'Trees', 'Wine', 'Coffee', 'Beans',
       'Fruits', 'Grains', 'Juices', 'Mushrooms', 'PetFood', 'Tofu',
       'WildHarvested']]

#Pandas can also plot multiple columns if the DataFrame includes them

multi_plot = Chart_df.plot(kind="bar", figsize=(15,13))
# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(Chart_df["city"], rotation=45)

```
