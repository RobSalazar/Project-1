```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
File=pd.read_csv("../clean_farm.csv")
File
File.fillna("N/A", axis=0, inplace=True)

NewFile=File[File['city'].isin([ "New York","San Francisco", "Los Angeles", "Seattle", "Boston"])]
```

```python
Products=NewFile.set_index('city')
```

```python
Products.columns
```

```python
Products.head()
```

```python
Products.columns[0]
```

```python
df=Products[['Organic', 'Bakedgoods',
       'Cheese','Flowers', 'Eggs', 'Seafood', 'Herbs', 'Vegetables',
       'Honey', 'Jams', 'Maple', 'Meat', 'Nuts', 'Plants',
       'Poultry', 'Prepared', 'Wine', 'Coffee', 'Beans',
       'Fruits', 'Grains', 'Juices', 'Mushrooms', 'PetFood', 'Tofu',
       'WildHarvested']]
df.head()
```

```python
ProductFile=df.reset_index()
ProductFile.head()
```

```python
Prodc_df=ProductFile.replace(["-", "Y", "N", "N/A"], [0, 1, 0, 0])
```

```python
Prodc_df.head()
```

```python
Sum_df=Prodc_df.groupby('city').sum()
```

```python
Total_Products=Sum_df.loc[:,'Total Product'] = Sum_df.sum(axis=1)
```

```python
Sum_df
```

```python
NewTotal=Sum_df.reset_index()
NewTotal
```

```python
NewTotal.plot(kind="bar", x="city", y="Total Product", grid=True, figsize=(8,8),
              title="Total Veriety Products")
```

```python
PVeriety=NewTotal[['city','Organic', 'Bakedgoods',
       'Cheese','Flowers', 'Eggs', 'Seafood', 'Herbs', 'Vegetables',
       'Honey', 'Jams', 'Maple', 'Meat', 'Nuts', 'Plants',
       'Poultry', 'Prepared', 'Wine', 'Coffee', 'Beans',
       'Fruits', 'Grains', 'Juices', 'Mushrooms', 'PetFood', 'Tofu',
       'WildHarvested']]

PVeriety
```

```python
PVeriety_df=PVeriety.transpose()
```

```python
PrtVeriety=PVeriety_df.rename(columns=PVeriety_df.iloc[0]).drop(PVeriety_df.index[0])
PrtVeriety_df=PrtVeriety.reset_index()
PrtVeriety_df.head()
```

```python
PrtVeriety.plot(figsize=[12, 7])
plt.legend(loc='upper right')
plt.xlabel("Products", fontsize=15)
plt.ylabel("Counts", fontsize=15)
plt.show()
```

```python
ProdtVeriety=PrtVeriety.mean()
ProdtVeriety
Cities=ProdtVeriety.keys()
Cities
```
